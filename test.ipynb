Podem fer servir una altra llibreria de python: pandas, per dibuixar una taula

Un grup de persones surten amb diferents mitjants de locomoció, de diferents punts i a diferents velocitats

```python
import pandas as pd

df = pd.DataFrame({
    'nom':['alfons', 'léa', 'chloé', 'irene'],
    'es mou':['a peu','en bici','en bus','en cotxe'],
    'velocitat en Km/h':[5,30,45,100],
    'distancia de origen en Km':[0,-5,-20,-80],
    })
df[['nom','es mou','velocitat en Km/h','distancia de origen en Km']]
```

Com sabem que la distancia és una funcio lineal en el temps $d=v*t$ podem fer una representació gràfica

```python
fig, ax = plt.subplots()
x = np.linspace(0, 2, 100)
ax.plot(x, 5*x, color="red", label="alfons, a peu")
ax.plot(x, 30*x-5, color="blue", label="léa, en bici")
ax.plot(x, 45*x-20, color="green",label="chloé, en bus")
ax.plot(x,100*x-80, color="purple",label="irene, en cotxe")
fig.tight_layout()
ax.legend(loc=2) # upper left corner
ax.set_xlabel(r'$hores$', fontsize=18)
ax.set_ylabel(r'$ Km $', fontsize=18)
ax.set_ylim(0,50);
```

Els punts de tall ens indiquen en quin moment i lloc es troben aquestes persones.

podriem haver disposat d'un altre estil de taula, per exemple de la que ha anat anotant la persona que anava a peu, on ha apuntat, la distancia que havia recorregut, a mesura que passava el temps:

```python
df = pd.DataFrame({
    'temps en h':[1, 2, 3, 4, 5],
    'distancia en Km':[5.2,9.8,14.7,18.1,23.2],
     })
df[['temps en h','distancia en Km']]
```

I ara fer servir el matplotlib per fer una grafica d'aquesta taula

```python
df.plot(kind='scatter',x='temps en h',y='distancia en Km',color='red');
```

La pendent de la recta ens dona la velocitat a la que s'ha mogut la persona

Per calcular la pendent farem una regressió lineal amb les dades, fent servir la funció scipy.stats.linregress(), aixi que primer hem d'importar stats

```python
from scipy import stats
```

Per poder fer una regressio lineal amb les dades del dataframe de pandas, primer l'hem de convertir en un array, per entendre-ho mirem com és per dins el dataframe:

```python
X = df[['temps en h']]
Y = df[['distancia en Km']]
print(X)
print(Y)
```

fem servir la funcio df.to_numpy() per importar el DataFrame en un vector de 2 dimensions

```python
Z = df.to_numpy()
print(Z)
```

I ara assignem les variables X i Y a cadascun dels vectors

```python
X=Z[:,0]
Y=Z[:,1]
print(X)
print(Y)
```

Ja estem en condicions de fer la regressio lineal amb les dades

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(X, Y)
print("pendent: %f    R-squared: %f" % (slope, r_value**2))
```

La pendent ens dona la velocitat en Km/h i R-squared ens diu com de bona és la recta (quan més a prop de 1 millor)
